```python
# CELL 1 — Load and Prepare IoT-23 Zeek Flow Dataset

import pandas as pd
import os
import glob

BASE_DIR = "/content/IoT23"

# 1) Locate all official Zeek-labeled logs
conn_files = glob.glob(os.path.join(BASE_DIR, "**", "bro", "conn.log.labeled"), recursive=True)
print("Found labeled files:")
for f in conn_files:
    print(" -", f)

# 2) Zeek conn.log parser
def load_zeek_conn(path):
    fields = None
    with open(path, "r", errors="ignore") as f:
        for line in f:
            if line.startswith("#fields"):
                fields = line.strip().split()[1:]
                break

    if fields is None:
        fields = [
            "ts","uid","id.orig_h","id.orig_p","id.resp_h","id.resp_p",
            "proto","service","duration","orig_bytes","resp_bytes",
            "orig_pkts","resp_pkts","conn_state","label"
        ]

    return pd.read_csv(path, sep="\t", comment="#", names=fields, header=None, low_memory=False)

# 3) Scenario-level ML label assignment
def assign_label_from_scenario(scenario):
    scen = scenario.lower()
    if "benign" in scen: return 0
    if "scan"   in scen: return 1
    if "cnc"    in scen: return 2
    return 2

# 4) Load + prepare all scenarios
all_data = []

for path in conn_files:
    scenario = path.split("/")[-3]
    print(f"\nLoading scenario: {scenario}")

    df = load_zeek_conn(path)

    keep = [
        "ts","id.orig_h","id.orig_p","id.resp_h","id.resp_p","proto","service",
        "duration","orig_bytes","resp_bytes","orig_pkts","resp_pkts","conn_state","label"
    ]
    df = df[[c for c in keep if c in df.columns]].copy()

    # Convert numeric features
    for col in ["duration","orig_bytes","resp_bytes","orig_pkts","resp_pkts"]:
        df[col] = pd.to_numeric(df[col], errors="coerce")  # fixed warning

    df["scenario"] = scenario
    df["ML_label"] = df["scenario"].apply(assign_label_from_scenario)

    print(f"Rows used: {len(df)}")
    all_data.append(df)

# 5) Merge
final_df = pd.concat(all_data, ignore_index=True)
print("\nFINAL MERGED SHAPE:", final_df.shape)

# 6) Save flows
final_df.to_csv("/content/flows.csv", index=False)
print("Saved flows to /content/flows.csv")

final_df.head()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical

# ==========================================
# 1. LOAD DATA & FEATURE ENGINEERING (Same as XGBoost)
# ==========================================
print("Loading flows...")
df = pd.read_csv("/content/flows.csv", low_memory=False)

# Clean 'service'
df['service'] = df['service'].fillna('unknown').replace('-', 'unknown')

# Clean numeric columns
numeric_cols = ["duration", "orig_bytes", "resp_bytes", "orig_pkts", "resp_pkts"]
df[numeric_cols] = df[numeric_cols].fillna(0)

# Clean Ports
df['id.resp_p'] = pd.to_numeric(df['id.resp_p'], errors='coerce').fillna(-1)

# Create Ratios
df['bytes_ratio'] = df['resp_bytes'] / (df['orig_bytes'] + 1)
df['pkts_ratio'] = df['resp_pkts'] / (df['orig_pkts'] + 1)

# Encode Categorical Data
df = pd.get_dummies(df, columns=['proto', 'conn_state'], drop_first=True)
le_service = LabelEncoder()
df['service_enc'] = le_service.fit_transform(df['service'].astype(str))

# Prepare Features
drop_cols = ['ts', 'uid', 'id.orig_h', 'id.orig_p', 'id.resp_h', 'label', 'scenario', 'ML_label', 'service']
features = [c for c in df.columns if c not in drop_cols]
X = df[features].values
y = df["ML_label"].values # 0=Benign, 1=Scan, 2=CnC

# Scale Features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split Global Test Set (We need a clean test set to evaluate the final system)
X_train_full, X_test, y_train_full, y_test = train_test_split(
    X_scaled, y, test_size=0.20, random_state=42, stratify=y
)

print(f"Data Loaded. Training Shape: {X_train_full.shape}")

# ==========================================
# 2. CREATE MODULAR DATASETS (The Strategy)
# ==========================================
# We split the training data into specific pairs for each module

def get_binary_data(X, y, class_a, class_b):
    # Filter only rows belonging to class_a or class_b
    mask = (y == class_a) | (y == class_b)
    X_sub = X[mask]
    y_sub = y[mask]
    # Remap labels: class_a -> 0, class_b -> 1 (Binary)
    y_sub_binary = np.where(y_sub == class_b, 1, 0)
    return X_sub, y_sub_binary

print("\nPreparing Modular Datasets...")

# Module 01: Benign (0) vs Scan (1)
X_01, y_01 = get_binary_data(X_train_full, y_train_full, 0, 1)

# Module 02: Benign (0) vs CnC (2)
X_02, y_02 = get_binary_data(X_train_full, y_train_full, 0, 2)

# Module 12: Scan (1) vs CnC (2) -> THE CRITICAL MODULE
X_12, y_12 = get_binary_data(X_train_full, y_train_full, 1, 2)

print(f"Module 01 (Benign vs Scan) samples: {len(y_01)}")
print(f"Module 02 (Benign vs CnC)  samples: {len(y_02)}")
print(f"Module 12 (Scan vs CnC)    samples: {len(y_12)}")

# ==========================================
# 3. BUILD & TRAIN MODULES
# ==========================================

def build_module(input_dim, name):
    model = Sequential(name=name)
    model.add(Dense(64, activation='relu', input_dim=input_dim))
    model.add(Dropout(0.2))
    model.add(Dense(32, activation='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(1, activation='sigmoid')) # Binary Output
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

print("\nTraining Module 01 (Benign vs Scan)...")
m01 = build_module(X.shape[1], "M01")
m01.fit(X_01, y_01, epochs=10, batch_size=64, verbose=0)

print("Training Module 02 (Benign vs CnC)...")
m02 = build_module(X.shape[1], "M02")
m02.fit(X_02, y_02, epochs=10, batch_size=64, verbose=0)

print("Training Module 12 (Scan vs CnC)...")
m12 = build_module(X.shape[1], "M12")
m12.fit(X_12, y_12, epochs=15, batch_size=64, verbose=0) # More epochs for the hard one

print("All Modules Trained.")

# ==========================================
# 4. MODULAR INFERENCE & COMBINATION
# ==========================================
print("\nRunning Inference...")

# Get raw probabilities from each specialist
# p01: High means Scan, Low means Benign
# p02: High means CnC, Low means Benign
# p12: High means CnC, Low means Scan
p01 = m01.predict(X_test, verbose=0)
p02 = m02.predict(X_test, verbose=0)
p12 = m12.predict(X_test, verbose=0)

# LOGIC FROM THE PAPER (Min-Max Combination)
# To be class 0 (Benign), you must win against Scan (1-p01) AND win against CnC (1-p02)
score_0 = np.minimum((1 - p01), (1 - p02))

# To be class 1 (Scan), you must win against Benign (p01) AND win against CnC (1-p12)
score_1 = np.minimum(p01, (1 - p12))

# To be class 2 (CnC), you must win against Benign (p02) AND win against Scan (p12)
score_2 = np.minimum(p02, p12)

# Stack scores side by side: [N_samples, 3]
final_scores = np.hstack([score_0, score_1, score_2])

# The winner is the one with the highest confidence "Min" score
y_pred = np.argmax(final_scores, axis=1)

# ==========================================
# 5. EVALUATION
# ==========================================
acc = accuracy_score(y_test, y_pred)
print(f"\nMODULAR NN Accuracy: {acc*100:.2f}%")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Purples",
            xticklabels=["Benign","Scan","CnC"],
            yticklabels=["Benign","Scan","CnC"])
plt.title(f"Modular Neural Network (Acc: {acc:.4f})")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
# CELL 5 — Three binary modules (XGBoost + small DNN)
# Modules: Benign-vs-others, Scan-vs-others, CnC-vs-others

# Requirements: X_train, X_test, y_train, y_test, scaler must exist from previous cells

# 1) Imports
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Input
from tensorflow.keras.callbacks import EarlyStopping

# Utility to plot confusion matrix
def plot_cm(cm, title):
    plt.figure(figsize=(4,3))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
    plt.title(title)
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

# Map multi-class labels to binary for a target class
def make_binary_labels(y, target_class):
    # returns array: 1 for target_class, 0 otherwise
    return np.where(y == target_class, 1, 0)

# Keep track of trained models and outputs
modules = ["Benign", "Scan", "CnC"]
target_map = {"Benign":0, "Scan":1, "CnC":2}

xgb_modules = {}
dnn_modules = {}
probs_train = {}  # store module probabilities on train (for combiner later if needed)
probs_test = {}   # store module probabilities on test

# Use a small DNN builder function
def build_small_dnn(input_shape):
    model = Sequential([
        Input(shape=(input_shape,)),
        Dense(64, activation='relu'),
        Dropout(0.2),
        Dense(32, activation='relu'),
        Dropout(0.1),
        Dense(1, activation='sigmoid')  # binary output
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Training loop for each module
for mod in modules:
    print(f"\n\n===== Module: {mod} (binary) =====")
    tclass = target_map[mod]

    # Prepare binary labels
    y_train_bin = make_binary_labels(y_train, tclass)
    y_test_bin  = make_binary_labels(y_test, tclass)

    # --- XGBoost (fast, strong baseline) ---
    print("\n-> Training XGBoost...")
    xgb = XGBClassifier(
        n_estimators=150,
        max_depth=5,
        learning_rate=0.1,
        use_label_encoder=False,
        eval_metric='logloss',
        random_state=42
    )
    xgb.fit(X_train, y_train_bin)
    y_pred_xgb = xgb.predict(X_test)
    y_prob_xgb = xgb.predict_proba(X_test)[:,1]

    print("XGBoost metrics:")
    print(" Accuracy:", accuracy_score(y_test_bin, y_pred_xgb))
    print(classification_report(y_test_bin, y_pred_xgb, digits=4))
    cm_x = confusion_matrix(y_test_bin, y_pred_xgb)
    plot_cm(cm_x, f"XGB CM - {mod}")

    # Save
    xgb_modules[mod] = xgb
    probs_test[f"xgb_{mod}"] = y_prob_xgb

    # --- Small DNN (Keras) ---
    print("\n-> Training small DNN...")
    dnn = build_small_dnn(X_train.shape[1])
    # Early stopping to avoid long runs
    es = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True, verbose=0)
    dnn.fit(X_train, y_train_bin, epochs=30, batch_size=128, validation_split=0.15, callbacks=[es], verbose=0)

    y_prob_dnn = dnn.predict(X_test).flatten()
    y_pred_dnn = (y_prob_dnn >= 0.5).astype(int)

    print("DNN metrics:")
    print(" Accuracy:", accuracy_score(y_test_bin, y_pred_dnn))
    print(classification_report(y_test_bin, y_pred_dnn, digits=4))
    cm_d = confusion_matrix(y_test_bin, y_pred_dnn)
    plot_cm(cm_d, f"DNN CM - {mod}")

    # Save
    dnn_modules[mod] = dnn
    probs_test[f"dnn_{mod}"] = y_prob_dnn

# Summarize stored probability keys
print("\nStored module probability keys (test):", list(probs_test.keys()))
```

```python
# CELL 6 — Adaptive Modular Neural Network (AMNN) combiner (hybrid stacking)
# Uses module probabilities from XGBoost + DNN modules (from CELL 5)

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1) Build module probability features for a dataset selector (train/val/test)
def build_module_probs(X, xgb_models, dnn_models):
    """
    X: (N, F) scaled numeric features
    returns: (N, 6) array of probabilities:
    [xgb_benign, xgb_scan, xgb_cnc, dnn_benign, dnn_scan, dnn_cnc]
    """
    probs = []
    # XGBoost module probs: predict_proba -> we used binary tasks, so idx 1 = prob of positive (target)
    p_xgb_b = xgb_models["Benign"].predict_proba(X)[:,1]
    p_xgb_s = xgb_models["Scan"].predict_proba(X)[:,1]
    p_xgb_c = xgb_models["CnC"].predict_proba(X)[:,1]
    # DNN models: they output a single probability per sample
    p_dnn_b = dnn_models["Benign"].predict(X).flatten()
    p_dnn_s = dnn_models["Scan"].predict(X).flatten()
    p_dnn_c = dnn_models["CnC"].predict(X).flatten()
    probs = np.vstack([p_xgb_b, p_xgb_s, p_xgb_c, p_dnn_b, p_dnn_s, p_dnn_c]).T
    return probs

# Build meta-features for TRAIN and TEST
meta_X_train_full = build_module_probs(X_train, xgb_modules, dnn_modules)
meta_X_test = build_module_probs(X_test, xgb_modules, dnn_modules)
meta_y_train_full = y_train  # same labels

print("Meta-features shapes:", meta_X_train_full.shape, meta_X_test.shape)

# 2) Split meta training into train_meta and val_meta to tune/log performance
from sklearn.model_selection import train_test_split
meta_X_train, meta_X_val, meta_y_train, meta_y_val = train_test_split(
    meta_X_train_full, meta_y_train_full, test_size=0.20, random_state=42, stratify=meta_y_train_full
)
print("Meta train/val shapes:", meta_X_train.shape, meta_X_val.shape)

# 3) Baseline 1: simple argmax of XGBoost module probs -> predicted class by argmax of [p_b, p_s, p_c]
def argmax_from_xgb(X_meta):
    p_xgb_only = X_meta[:,0:3]  # first 3 are xgb probs
    preds = np.argmax(p_xgb_only, axis=1)  # 0->Benign,1->Scan,2->CnC
    return preds

# Baseline 2: argmax of DNN module probs
def argmax_from_dnn(X_meta):
    p_dnn_only = X_meta[:,3:6]
    preds = np.argmax(p_dnn_only, axis=1)
    return preds

# Evaluate baselines on test set
pred_xgb_baseline = argmax_from_xgb(meta_X_test)
pred_dnn_baseline = argmax_from_dnn(meta_X_test)

print("\nBaseline XGB-modules performance on test:")
print("Accuracy:", accuracy_score(y_test, pred_xgb_baseline))
print(classification_report(y_test, pred_xgb_baseline, digits=4))

print("\nBaseline DNN-modules performance on test:")
print("Accuracy:", accuracy_score(y_test, pred_dnn_baseline))
print(classification_report(y_test, pred_dnn_baseline, digits=4))

# 4) Train meta-classifier (Logistic Regression multinomial)
meta_clf = LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=2000, random_state=42)
meta_clf.fit(meta_X_train, meta_y_train)

# Evaluate on validation set
val_pred = meta_clf.predict(meta_X_val)
print("\nMeta-classifier (val) Accuracy:", accuracy_score(meta_y_val, val_pred))
print(classification_report(meta_y_val, val_pred, digits=4))

# Evaluate final on test set
meta_test_pred = meta_clf.predict(meta_X_test)
print("\nAMNN (meta) on test Accuracy:", accuracy_score(y_test, meta_test_pred))
print(classification_report(y_test, meta_test_pred, digits=4))

# Confusion matrix for meta-combiner
cm_meta = confusion_matrix(y_test, meta_test_pred)
plt.figure(figsize=(6,4))
sns.heatmap(cm_meta, annot=True, fmt='d', cmap='Oranges',
            xticklabels=["Benign","Scan","CnC"], yticklabels=["Benign","Scan","CnC"])
plt.title("AMNN (meta) Confusion Matrix")
plt.xlabel("Predicted"); plt.ylabel("Actual")
plt.show()

# 5) Save meta model (optional) and show feature importance (coeffs)
import joblib
joblib.dump(meta_clf, "/content/amnn_meta_clf.joblib")
print("Saved meta-classifier to /content/amnn_meta_clf.joblib")

# Show weights learned (logistic regression coefficients)
coef = meta_clf.coef_  # shape (3,6)
print("\nMeta-classifier coefficients (rows = class 0/1/2, cols = xgb_b,xgb_s,xgb_c,dnn_b,dnn_s,dnn_c):")
print(coef)

# 6) Quick comparison table
from sklearn.metrics import f1_score, precision_score, recall_score
def summarize(name, y_true, y_pred):
    return {
        "name": name,
        "accuracy": accuracy_score(y_true, y_pred),
        "f1_macro": f1_score(y_true, y_pred, average='macro'),
        "precision_macro": precision_score(y_true, y_pred, average='macro'),
        "recall_macro": recall_score(y_true, y_pred, average='macro')
    }

summary = [
    summarize("XGB-module-argmax", y_test, pred_xgb_baseline),
    summarize("DNN-module-argmax", y_test, pred_dnn_baseline),
    summarize("AMNN-meta", y_test, meta_test_pred)
]

import pandas as pd
print("\nSummary comparison:")
pd.DataFrame(summary)
```

```python
# CELL 8 — Unified AMNN Inference Pipeline


import joblib
import numpy as np
import tensorflow as tf

# Paths to saved models
BASE = "/content/saved_models"

# Load scaler
scaler_loaded = joblib.load(f"{BASE}/scaler/standard_scaler.pkl")

# Load XGB modules
xgb_benign = joblib.load(f"{BASE}/xgb_modules/xgb_Benign.pkl")
xgb_scan   = joblib.load(f"{BASE}/xgb_modules/xgb_Scan.pkl")
xgb_cnc    = joblib.load(f"{BASE}/xgb_modules/xgb_CnC.pkl")

# Load DNN modules
dnn_benign = tf.keras.models.load_model(f"{BASE}/dnn_modules/dnn_Benign.h5")
dnn_scan   = tf.keras.models.load_model(f"{BASE}/dnn_modules/dnn_Scan.h5")
dnn_cnc    = tf.keras.models.load_model(f"{BASE}/dnn_modules/dnn_CnC.h5")

# Load meta-classifier
meta_clf_loaded = joblib.load(f"{BASE}/amnn/meta_classifier.pkl")

print("All models loaded successfully!")


# ----------------------------------------------------
# FINAL AMNN PREDICTION FUNCTION
# ----------------------------------------------------
def amnn_predict(flow_dict):
    """
    flow_dict must contain the 5 numerical features:
      duration, orig_bytes, resp_bytes, orig_pkts, resp_pkts
    Example:
      {"duration": 1.2, "orig_bytes": 50, "resp_bytes": 300,
       "orig_pkts": 10, "resp_pkts": 12}
    Returns:
      final_class (0=Benign, 1=Scan, 2=CnC)
    """

    # 1) Convert dict → numpy array
    features = np.array([
        flow_dict["duration"],
        flow_dict["orig_bytes"],
        flow_dict["resp_bytes"],
        flow_dict["orig_pkts"],
        flow_dict["resp_pkts"]
    ]).reshape(1, -1)

    # 2) Scale
    X_scaled = scaler_loaded.transform(features)

    # 3) Get probabilities from XGB modules
    p_xgb_b = xgb_benign.predict_proba(X_scaled)[:,1]
    p_xgb_s = xgb_scan.predict_proba(X_scaled)[:,1]
    p_xgb_c = xgb_cnc.predict_proba(X_scaled)[:,1]

    # 4) Get probabilities from DNN modules
    p_dnn_b = dnn_benign.predict(X_scaled).flatten()
    p_dnn_s = dnn_scan.predict(X_scaled).flatten()
    p_dnn_c = dnn_cnc.predict(X_scaled).flatten()

    # 5) Build meta-feature vector
    meta_vec = np.array([
        p_xgb_b[0], p_xgb_s[0], p_xgb_c[0],
        p_dnn_b[0], p_dnn_s[0], p_dnn_c[0]
    ]).reshape(1, -1)

    # 6) Meta-classifier final prediction
    final_class = meta_clf_loaded.predict(meta_vec)[0]

    return final_class



# TEST THE FINAL PIPELINE WITH A SAMPLE INPUT

sample_flow = {
    "duration": 1.5,
    "orig_bytes": 100,
    "resp_bytes": 200,
    "orig_pkts": 5,
    "resp_pkts": 7
}

print("Sample Prediction:", amnn_predict(sample_flow))
```

```python
# Colab-ready cell: MNN implementation for IoT-23
# Paste and run. Requires: pip install imbalanced-learn xgboost (optional)
!pip install -q imbalanced-learn xgboost

import os
import random
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import tensorflow as tf
from tensorflow.keras import layers, models, callbacks, losses, optimizers
from imblearn.over_sampling import SMOTE
import joblib
import matplotlib.pyplot as plt
import seaborn as sns

# ---------- CONFIG ----------
DATA_PATH = "/content/conn.log.labeled"   # change to your IoT-23 CSV path
RANDOM_SEEDS = [42, 2024, 7]               # ensemble over seeds
RESULTS_DIR = "/content/mnn_results"
os.makedirs(RESULTS_DIR, exist_ok=True)

# Module classes (labels in IoT-23 that you have to map to these three categories)
# Adjust label mapping as needed depending on your conn.log.labeled label names
BENIGN_LABELS = ["Benign"]               # or "benign" depending on your file
SCAN_LABELS = ["Scan"]                   # adjust
CNC_LABELS = ["C&C", "CnC", "Malicious"] # adjust

# ---------- UTILITIES ----------
def set_global_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    tf.random.set_seed(seed)

def read_iot23(path):
    """Read IoT-23 preprocessed CSV. If your conn.log.labeled has a different layout,
       adapt this function to extract the same features used in the paper.
    """
    df = pd.read_csv(path)
    # Expect a column 'label' with class names - adjust if different
    if 'label' not in df.columns and 'Label' in df.columns:
        df.rename(columns={'Label': 'label'}, inplace=True)
    return df

def prepare_features(df, feature_cols=None, scaler=None, fit_scaler=True):
    if feature_cols is None:
        # Heuristic: drop non-numeric cols, drop label
        feature_cols = [c for c in df.columns if df[c].dtype in [np.float64, np.int64]]
        # If 'label' is numeric, remove it
        feature_cols = [c for c in feature_cols if c.lower() not in ('label','labels','class')]
    X = df[feature_cols].astype(float).fillna(0.0).values
    if scaler is None:
        scaler = StandardScaler()
    if fit_scaler:
        X = scaler.fit_transform(X)
    else:
        X = scaler.transform(X)
    return X, scaler, feature_cols

def map_to_three_labels(y_raw):
    """Map dataset labels to 'Benign', 'Scan', 'CnC' strings.
       Modify mapping rules according to your file's label names.
    """
    y3 = []
    for v in y_raw:
        s = str(v).lower()
        if 'benign' in s:
            y3.append('Benign')
        elif 'scan' in s or 'scanning' in s:
            y3.append('Scan')
        else:
            # remaining and explicit 'cnc' or 'c&c' etc
            y3.append('C&C')
    return np.array(y3)

# ---------- MODEL: small binary DNN for a module ----------
def build_module_model(input_dim, hidden_sizes=[128,64], dropout=0.2):
    inp = layers.Input(shape=(input_dim,))
    x = inp
    for h in hidden_sizes:
        x = layers.Dense(h, activation='relu')(x)
        x = layers.BatchNormalization()(x)
        x = layers.Dropout(dropout)(x)
    out = layers.Dense(1, activation='sigmoid')(x)
    model = models.Model(inputs=inp, outputs=out)
    model.compile(optimizer=optimizers.Adam(learning_rate=1e-3),
                  loss=losses.BinaryCrossentropy(),
                  metrics=['AUC'])
    return model

# Optional focal loss for severe class imbalance:
def focal_loss(gamma=2., alpha=.25):
    def fl(y_true, y_pred):
        y_pred = tf.clip_by_value(y_pred, 1e-7, 1-1e-7)
        pt = tf.where(tf.equal(y_true,1), y_pred, 1-y_pred)
        w = tf.where(tf.equal(y_true,1), alpha, (1-alpha))
        return -w * tf.math.pow(1-pt, gamma) * tf.math.log(pt)
    return fl

# ---------- TRAIN & EVAL ----------
def train_binary_module(X_train, y_train, X_val, y_val, seed=42, use_focal=False, smote=False):
    set_global_seed(seed)
    if smote:
        sm = SMOTE(random_state=seed)
        X_train, y_train = sm.fit_resample(X_train, y_train)
    model = build_module_model(X_train.shape[1])
    if use_focal:
        model.compile(optimizer=optimizers.Adam(1e-3), loss=focal_loss(), metrics=['AUC'])
    es = callbacks.EarlyStopping(monitor='val_loss', patience=6, restore_best_weights=True)
    hist = model.fit(X_train, y_train, validation_data=(X_val, y_val),
                     epochs=100, batch_size=1024, callbacks=[es], verbose=0)
    return model, hist

def module_predict_proba(model, X):
    return model.predict(X, batch_size=4096).ravel()

# ---------- END-TO-END RUN for one seed ----------
def run_mnn_for_seed(df, seed=42, use_smote_on_cnc=False, use_focal=False):
    set_global_seed(seed)
    # Map to 3-class labels
    df2 = df.copy()
    df2['label3'] = map_to_three_labels(df2['label'].values)
    # Train/val/test split (stratified)
    X_full, scaler, feat_cols = prepare_features(df2)
    le = LabelEncoder()
    y3 = le.fit_transform(df2['label3'].values)  # 0:Benign,1:CnC,2:Scan possibly order differs - check mapping
    # Use stratified split
    X_train, X_temp, y_train, y_temp, lbl_train, lbl_temp = train_test_split(X_full, y3, df2['label3'].values,
                                                                             test_size=0.4, random_state=seed,
                                                                             stratify=y3)
    X_val, X_test, y_val, y_test, lbl_val, lbl_test = train_test_split(X_temp, y_temp, lbl_temp,
                                                                       test_size=0.5, random_state=seed,
                                                                       stratify=y_temp)
    # Save scaler & feature columns for reproducibility
    joblib.dump(scaler, os.path.join(RESULTS_DIR, f"scaler_seed{seed}.pkl"))
    pd.DataFrame({'feature': feat_cols}).to_csv(os.path.join(RESULTS_DIR, f"features_seed{seed}.csv"), index=False)

    # Build binary labels for modules: module outputs prob(class)
    # We'll create three modules: Benign, Scan, C&C each "is class vs rest"
    label_names = ['Benign','Scan','C&C']
    module_models = {}
    module_val_preds = {}
    module_test_preds = {}
    for cls in label_names:
        y_train_bin = (lbl_train == cls).astype(int)
        y_val_bin   = (lbl_val   == cls).astype(int)
        y_test_bin  = (lbl_test  == cls).astype(int)
        smote = (use_smote_on_cnc and cls in ('C&C',))
        model, hist = train_binary_module(X_train, y_train_bin, X_val, y_val_bin,
                                          seed=seed, use_focal=use_focal, smote=smote)
        module_models[cls] = model
        module_val_preds[cls] = module_predict_proba(model, X_val)
        module_test_preds[cls] = module_predict_proba(model, X_test)
        # save model
        model.save(os.path.join(RESULTS_DIR, f"module_{cls}_seed{seed}.keras"))

    # Final decision rule: take argmax over module probabilities for each sample
    proba_matrix_test = np.vstack([module_test_preds[c] for c in label_names]).T
    pred_idx = np.argmax(proba_matrix_test, axis=1)
    pred_labels = [label_names[i] for i in pred_idx]

    # Evaluate
    acc = accuracy_score(lbl_test, pred_labels)
    rep = classification_report(lbl_test, pred_labels, output_dict=False)
    cm = confusion_matrix(lbl_test, pred_labels, labels=label_names)

    # Save outputs
    np.save(os.path.join(RESULTS_DIR, f"proba_test_seed{seed}.npy"), proba_matrix_test)
    pd.DataFrame(cm, index=label_names, columns=label_names).to_csv(os.path.join(RESULTS_DIR, f"cm_seed{seed}.csv"))
    with open(os.path.join(RESULTS_DIR, f"report_seed{seed}.txt"), 'w') as f:
        f.write(f"Accuracy: {acc}\n\n")
        f.write(rep)

    return dict(seed=seed, acc=acc, report=rep, cm=cm, labels=label_names)

# ---------- MASTER RUN over seeds (ensemble averaging) ----------
def master_run(data_path=DATA_PATH, seeds=RANDOM_SEEDS, use_smote_on_cnc=False, use_focal=False):
    df = read_iot23(data_path)
    results = []
    for s in seeds:
        print("Running seed:", s)
        res = run_mnn_for_seed(df, seed=s, use_smote_on_cnc=use_smote_on_cnc, use_focal=use_focal)
        print("Seed", s, "acc:", res['acc'])
        results.append(res)
    # Aggregate
    accs = [r['acc'] for r in results]
    print("Acc per seed:", accs)
    print("Mean acc: %.4f, std: %.4f" % (np.mean(accs), np.std(accs)))
    # Optionally produce combined confusion matrix by summing per-seed confusion matrices (or majority voting)
    cms = [r['cm'] for r in results]
    cm_sum = np.sum(cms, axis=0)
    labels = results[0]['labels']
    pd.DataFrame(cm_sum, index=labels, columns=labels).to_csv(os.path.join(RESULTS_DIR, "cm_aggregate.csv"))
    # Plot aggregate confusion matrix
    plt.figure(figsize=(6,5))
    sns.heatmap(cm_sum, annot=True, fmt='d', xticklabels=labels, yticklabels=labels)
    plt.title("Aggregate confusion matrix (sum over seeds)")
    plt.ylabel("True")
    plt.xlabel("Pred")
    plt.savefig(os.path.join(RESULTS_DIR, "cm_aggregate.png"))
    plt.show()
    return results

# ---------- RUN ----------
# Make sure your conn.log.labeled has a 'label' column. If not, adjust read/write.
results = master_run(DATA_PATH, seeds=RANDOM_SEEDS, use_smote_on_cnc=True, use_focal=False)
```

```python
# ==========================================
# FINAL MNN IMPLEMENTATION (2-CLASS)
# Benign vs Malicious (Botnet)
# ==========================================

import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import SMOTE
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers, callbacks
import seaborn as sns
import matplotlib.pyplot as plt

# =============================
# 1. LOAD DATA IN STREAM MODE
# =============================

print("Loading dataset in chunks (RAM-safe)...")

chunks = []
reader = pd.read_csv("/content/iot23_final.csv", chunksize=200000)  # safe chunk

for chunk in reader:
    # Keep only numeric columns + label3
    if "label3" not in chunk.columns:
        continue
    chunk = chunk.select_dtypes(include=[np.number]).assign(label3=chunk["label3"])
    chunks.append(chunk)

df = pd.concat(chunks, ignore_index=True)
print("Loaded:", df.shape)

# =============================
# 2. PREPARE LABELS (Binary)
# =============================

df = df[ df["label3"].isin(["Benign", "C&C"]) ]  # ensure only 2 classes
df["label_binary"] = df["label3"].map({"Benign":0, "C&C":1})

X = df.drop(columns=["label3","label_binary"])
y = df["label_binary"]
print("Benign:", (y==0).sum(), "Malicious:", (y==1).sum())

# ==============================================
# 3. BALANCING (Under-sample + SMOTE)
# ==============================================

print("Balancing dataset...")

# Step 1: Reduce extreme class imbalance
rus = RandomUnderSampler(sampling_strategy=0.5)
X_rus, y_rus = rus.fit_resample(X, y)

# Step 2: Apply SMOTE to boost minority class
sm = SMOTE()
X_bal, y_bal = sm.fit_resample(X_rus, y_rus)

print("Balanced shape:", X_bal.shape, "Labels:", np.bincount(y_bal))

# =============================
# 4. Train-Test Split
# =============================

X_train, X_test, y_train, y_test = train_test_split(
    X_bal, y_bal, test_size=0.25, random_state=42, stratify=y_bal)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test  = scaler.transform(X_test)

# =============================
# 5. DEFINE MNN MODULES
# =============================

def build_module(input_dim):
    model = models.Sequential([
        layers.Dense(128, activation="relu", input_shape=(input_dim,)),
        layers.BatchNormalization(),
        layers.Dropout(0.3),
        layers.Dense(64, activation="relu"),
        layers.BatchNormalization(),
        layers.Dropout(0.3),
        layers.Dense(1, activation="sigmoid")
    ])
    model.compile(
        optimizer=optimizers.Adam(1e-3),
        loss="binary_crossentropy",
        metrics=["accuracy"]
    )
    return model

input_dim = X_train.shape[1]

module_benign   = build_module(input_dim)
module_malicious = build_module(input_dim)

print("Training MNN modules...")

es = callbacks.EarlyStopping(patience=4, restore_best_weights=True)

# =============================
# 6. TRAIN MODULE 1: BENIGN
# =============================
y_benign_train = (y_train == 0).astype(int)
module_benign.fit(
    X_train, y_benign_train,
    validation_split=0.1,
    epochs=20,
    batch_size=4096,
    callbacks=[es],
    verbose=1
)

# =============================
# 7. TRAIN MODULE 2: MALICIOUS
# =============================
y_mal_train = (y_train == 1).astype(int)
module_malicious.fit(
    X_train, y_mal_train,
    validation_split=0.1,
    epochs=20,
    batch_size=4096,
    callbacks=[es],
    verbose=1
)

# =============================
# 8. FUSION (Final Prediction)
# =============================

p_ben = module_benign.predict(X_test).ravel()        # Probability of benign
p_mal = module_malicious.predict(X_test).ravel()     # Probability of malicious

# CHOOSE CLASS WITH HIGHER PROBABILITY
y_pred = np.where(p_mal > p_ben, 1, 0)

# =============================
# 9. EVALUATION
# =============================

print("\n=== FINAL MNN RESULTS ===")
acc = accuracy_score(y_test, y_pred)
print("Accuracy:", acc)

print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=["Benign", "Malicious"]))

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt="d", xticklabels=["Benign","Malicious"], yticklabels=["Benign","Malicious"])
plt.title("Confusion Matrix")
plt.show()

```
