```python
import os
import pandas as pd

def load_connlog_with_header(path):
    header_fields = None
    rows = []
    with open(path, "r", encoding="utf-8", errors="ignore") as f:
        for line in f:
            if line.startswith("#fields"):
                # header line: '#fields\tts\tuid\t...'
                # remove '#fields' and split remaining into column names
                header_fields = line.strip().split()[1:]
                continue
            if line.startswith("#"):
                continue
            line = line.rstrip("\n")
            # try tab split first (Zeek uses tabs)
            parts = line.split("\t")
            if header_fields is not None and len(parts) != len(header_fields):
                # fallback: whitespace split (some files may not use strict tabs)
                parts = line.split()
            rows.append(parts)

    if header_fields is None:
        raise ValueError(f"No #fields header found in {path}")

    # Validate row lengths and pad/truncate if needed
    cleaned_rows = []
    for i, r in enumerate(rows):
        if len(r) < len(header_fields):
            # pad with empty strings
            r = r + [""] * (len(header_fields) - len(r))
        elif len(r) > len(header_fields):
            # join extra trailing pieces into last column (label may contain spaces)
            # put the overflow back into the last column
            extra = r[len(header_fields)-1:]
            r = r[:len(header_fields)-1] + [" ".join(extra)]
        cleaned_rows.append(r)

    df = pd.DataFrame(cleaned_rows, columns=header_fields)
    return df

# load all files using the robust parser
base = "/content/IoT23"
paths = {
    "Benign1": f"{base}/Benign1/bro/conn.log.labeled",
    "Benign2": f"{base}/Benign2/bro/conn.log.labeled",
    "Scan1":   f"{base}/Scan1/bro/conn.log.labeled",
    "Scan2":   f"{base}/Scan2/bro/conn.log.labeled",
    "CnC1":    f"{base}/CnC1/bro/conn.log.labeled",
    "CnC2":    f"{base}/CnC2/bro/conn.log.labeled",
}

dfs = []
for name, path in paths.items():
    print("Loading:", name, "->", path)
    df = load_connlog_with_header(path)
    df["scenario"] = name
    print("  shape:", df.shape, "columns:", len(df.columns))
    dfs.append(df)

df_all = pd.concat(dfs, ignore_index=True)
print("\nMerged shape:", df_all.shape)
print("Columns:\n", df_all.columns.tolist())
df_all.head()
```

```python
# Cell 1 - imports and seed control

import os
import random
import numpy as np
import pandas as pd

# ML
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score
from sklearn.utils.class_weight import compute_class_weight

import xgboost as xgb
import tensorflow as tf
from tensorflow.keras import layers, models, callbacks, optimizers, regularizers

# Reproducibility
RSEED = 42
os.environ['PYTHONHASHSEED'] = str(RSEED)
np.random.seed(RSEED)
random.seed(RSEED)
tf.random.set_seed(RSEED)

print("Cell-1 complete: imports and seeds set.")
```

```python
# Cell 2 — Clean labels and convert to 2-class

df = df_all.copy()

# Normalize case
df["label"] = df["label"].str.strip().str.lower()

# Map to binary
df["label2"] = df["label"].map({
    "benign": 0,
    "malicious": 1
})

print("Value counts after mapping:")
print(df["label2"].value_counts())

# Drop NaN labels if any
df = df.dropna(subset=["label2"])
df["label2"] = df["label2"].astype(int)

print("\nFinal binary class balance:")
print(df["label2"].value_counts())

df.head()
```

```python
# Cell 3 — Clean features, remove non-numeric fields, convert to numeric

df_clean = df.copy()

# Replace '-' with NaN everywhere
df_clean = df_clean.replace("-", pd.NA)

# Columns to drop (string / identifiers / non-ML)
drop_cols = [
    "uid", "id.orig_h", "id.resp_h",
    "proto", "service", "conn_state", "history",
    "tunnel_parents", "label", "detailed-label",
    "scenario", "det_label"
]

df_clean = df_clean.drop(columns=drop_cols, errors="ignore")

# Convert all remaining columns except label2 to numeric
for col in df_clean.columns:
    if col != "label2":
        df_clean[col] = pd.to_numeric(df_clean[col], errors="coerce")

# Drop rows where duration or bytes completely missing (rare)
df_clean = df_clean.dropna(subset=["duration", "orig_bytes", "resp_bytes"])

print("Final columns:", df_clean.columns.tolist())
print("Shape after cleaning:", df_clean.shape)

df_clean.head()
```

```python
# Cell 4 — Stratified Split

from sklearn.model_selection import train_test_split

df_ml = df_clean.copy()

X = df_ml.drop(columns=["label2"])
y = df_ml["label2"].values

# 70% train, 15% validation, 15% test
X_temp, X_test, y_temp, y_test = train_test_split(
    X, y, test_size=0.15, stratify=y, random_state=42
)

X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=0.17647, stratify=y_temp, random_state=42
)

print("Shapes:")
print("Train:", X_train.shape)
print("Val:  ", X_val.shape)
print("Test: ", X_test.shape)

print("\nClass balance in train:")
print(pd.Series(y_train).value_counts())
```

```python
# Cell 5 — Standardize numeric features

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# fit scaler ONLY on training data
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled   = scaler.transform(X_val)
X_test_scaled  = scaler.transform(X_test)

print("Scaling complete.")
print("Train scaled shape:", X_train_scaled.shape)
```

```python
# Cell 6 — class weights

from sklearn.utils.class_weight import compute_class_weight
import numpy as np

classes = np.unique(y_train)
weights = compute_class_weight(
    class_weight='balanced',
    classes=classes,
    y=y_train
)

class_weights = dict(zip(classes, weights))

print("Class weights:", class_weights)
```

```python
# Cell 7 — XGBoost Baseline (CPU version)

import xgboost as xgb
from sklearn.metrics import accuracy_score, roc_auc_score, classification_report

# Convert to DMatrix format
dtrain = xgb.DMatrix(X_train_scaled, label=y_train)
dval   = xgb.DMatrix(X_val_scaled, label=y_val)

params = {
    'objective': 'binary:logistic',
    'eval_metric': 'auc',
    'tree_method': 'hist',   # CPU-friendly version
    'learning_rate': 0.05,
    'max_depth': 6,
    'subsample': 0.8,
    'colsample_bytree': 0.8,
    'seed': 42
}

evals = [(dtrain, 'train'), (dval, 'val')]

bst = xgb.train(
    params,
    dtrain,
    num_boost_round=2000,
    early_stopping_rounds=50,
    evals=evals,
    verbose_eval=50
)

# Predictions
y_val_proba = bst.predict(dval)
y_val_pred = (y_val_proba >= 0.5).astype(int)

print("\nXGBoost Validation Accuracy:", accuracy_score(y_val, y_val_pred))
print("XGBoost Validation AUC:", roc_auc_score(y_val, y_val_proba))

print("\nClassification Report:\n", classification_report(y_val, y_val_pred))
```

```python
# Cell 8 — Generate XGBoost probability features for stacking

# Predictions on all sets
y_train_proba = bst.predict(xgb.DMatrix(X_train_scaled))
y_val_proba   = bst.predict(xgb.DMatrix(X_val_scaled))
y_test_proba  = bst.predict(xgb.DMatrix(X_test_scaled))

# Add as a new feature column
import numpy as np

X_train_am = np.hstack([X_train_scaled, y_train_proba.reshape(-1, 1)])
X_val_am   = np.hstack([X_val_scaled,   y_val_proba.reshape(-1,  1)])
X_test_am  = np.hstack([X_test_scaled,  y_test_proba.reshape(-1, 1)])

print("AMNN input shapes:")
print("Train:", X_train_am.shape)
print("Val:  ", X_val_am.shape)
print("Test: ", X_test_am.shape)
```

```python
# Cell 9 — Build AMNN Model

import tensorflow as tf
from tensorflow.keras import layers, models, regularizers, callbacks, optimizers

input_dim = X_train_am.shape[1]

def build_amnn(input_dim, lr=1e-3, l2=1e-4):
    inp = layers.Input(shape=(input_dim,))

    x = layers.Dense(256, activation='relu', kernel_regularizer=regularizers.l2(l2))(inp)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.3)(x)

    x = layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(l2))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.25)(x)

    x = layers.Dense(64, activation='relu', kernel_regularizer=regularizers.l2(l2))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.2)(x)

    out = layers.Dense(1, activation='sigmoid')(x)

    model = models.Model(inputs=inp, outputs=out)
    model.compile(
        optimizer=optimizers.Adam(learning_rate=lr),
        loss='binary_crossentropy',
        metrics=['accuracy', tf.keras.metrics.AUC(name='auc')]
    )

    return model

amnn = build_amnn(input_dim)
amnn.summary()
```

```python
# Cell 10 — Train AMNN

es = callbacks.EarlyStopping(
    monitor='val_auc',
    mode='max',
    patience=12,
    restore_best_weights=True,
    verbose=1
)

rlr = callbacks.ReduceLROnPlateau(
    monitor='val_auc',
    mode='max',
    factor=0.5,
    patience=5,
    min_lr=1e-6,
    verbose=1
)

mc = callbacks.ModelCheckpoint(
    "amnn_best.h5",
    monitor='val_auc',
    mode='max',
    save_best_only=True,
    verbose=1
)

history = amnn.fit(
    X_train_am, y_train,
    validation_data=(X_val_am, y_val),
    epochs=200,
    batch_size=256,
    class_weight=class_weights,
    callbacks=[es, rlr, mc],
    verbose=2
)
```
