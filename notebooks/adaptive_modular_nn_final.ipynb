```python
# ============================================
# CELL 1 — Load IoT-23 labeled flows (simple version)
# ============================================

import pandas as pd
import os
import glob

# Base directory where your scenarios are stored
BASE_DIR = "/content/IoT23"

# 1) Find all correct labeled files inside bro/ folders
conn_files = glob.glob(os.path.join(BASE_DIR, "**", "bro", "conn.log.labeled"), recursive=True)

print("Found labeled files:")
for f in conn_files:
    print(" -", f)

# 2) A simple Zeek conn.log parser
def load_zeek_conn(path):
    """
    Zeek conn.log.labeled files use tab-separated format.
    The column names come from the #fields header line.
    """
    # Read the header to find #fields
    fields = None
    with open(path, "r", errors="ignore") as f:
        for line in f:
            if line.startswith("#fields"):
                fields = line.strip().split()[1:]  # remove "#fields"
                break

    # Fallback if something goes wrong
    if fields is None:
        fields = [
            "ts","uid","id.orig_h","id.orig_p","id.resp_h","id.resp_p",
            "proto","service","duration","orig_bytes","resp_bytes",
            "conn_state","local_orig","local_resp","missed_bytes","history",
            "orig_pkts","orig_ip_bytes","resp_pkts","resp_ip_bytes",
            "tunnel_parents","label"
        ]

    # Load file ignoring Zeek comment lines starting with '#'
    df = pd.read_csv(path, sep="\t", comment="#", names=fields, header=None, low_memory=False)
    return df

# 3) Map IoT-23 labels to ML classes (0 = Benign, 1 = Scan, 2 = CnC)
def map_label(text):
    text = str(text).lower()
    if "benign" in text:
        return 0
    if "scan" in text or "horizontalportscan" in text or "portscan" in text:
        return 1
    if "c&c" in text or "cnc" in text or "filedownload" in text or "torii" in text or "mirai" in text:
        return 2
    # default: treat unknown malware as class 2 (malicious)
    return 2

# 4) Load all scenarios into one dataframe
all_data = []

for file_path in conn_files:
    scenario = file_path.split("/")[-3]  # folder name (Benign1, Scan1, etc.)
    print(f"\nLoading scenario: {scenario}")

    df = load_zeek_conn(file_path)

    # Keep only essential columns
    keep_cols = [
        "ts","id.orig_h","id.orig_p","id.resp_h","id.resp_p",
        "proto","service","duration","orig_bytes","resp_bytes",
        "orig_pkts","resp_pkts","conn_state","label"
    ]

    df = df[[c for c in keep_cols if c in df.columns]].copy()

    # Convert numerics
    for col in ["duration","orig_bytes","resp_bytes","orig_pkts","resp_pkts"]:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors="coerce")

    # Add scenario name
    df["scenario"] = scenario

    # Apply ML label mapping
    df["ML_label"] = df["label"].apply(map_label)

    # Drop rows missing ML label
    before = len(df)
    df = df.dropna(subset=["ML_label"])
    print(f"Rows used: {len(df)} (dropped {before - len(df)})")

    all_data.append(df)

# 5) Merge everything
final_df = pd.concat(all_data, ignore_index=True)
print("\nFINAL MERGED SHAPE:", final_df.shape)

# 6) Save to CSV
final_df.to_csv("/content/flows.csv", index=False)
print("\nSaved merged flows to /content/flows.csv")

# Show preview
final_df.head()
```

```python
# Fix ML labels using scenario names
def assign_label_from_scenario(row):
    scen = row["scenario"].lower()
    if "benign" in scen:
        return 0   # Benign
    if "scan" in scen:
        return 1   # Scanning / probing
    if "cnc" in scen:
        return 2   # Command & Control
    return row["ML_label"]  # fallback

final_df["ML_label"] = final_df.apply(assign_label_from_scenario, axis=1)

# Save updated version
final_df.to_csv("/content/flows.csv", index=False)

print("Updated ML_label mapping saved!")
final_df.head()
```

```python
# ============================================
# CELL 2 — Feature Extraction + Normalization
# ============================================

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Load flows.csv from previous step
df = pd.read_csv("/content/flows.csv")

print("Original shape:", df.shape)

# Select numeric ML features (commonly used in IoT-23 studies)
features = ["duration", "orig_bytes", "resp_bytes", "orig_pkts", "resp_pkts"]

# If some numeric fields are missing (NaN), fill with 0
df[features] = df[features].fillna(0)

# Input (X) and Output labels (y)
X = df[features].values
y = df["ML_label"].values

# Normalize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train/Test split (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.20, random_state=42, stratify=y
)

print("Training shape:", X_train.shape)
print("Testing shape:", X_test.shape)
```

```python
# ============================================
# CELL 3 — XGBoost Model
# ============================================

!pip install xgboost > /dev/null

from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

# Train the model
xgb = XGBClassifier(
    n_estimators=200,
    max_depth=6,
    learning_rate=0.1,
    subsample=0.8,
    colsample_bytree=0.8,
    objective="multi:softmax",
    num_class=3,
    random_state=42
)

xgb.fit(X_train, y_train)

# Predictions
y_pred = xgb.predict(X_test)

# Evaluation
print("\nAccuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=["Benign","Scan","CnC"],
            yticklabels=["Benign","Scan","CnC"])
plt.title("XGBoost Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
# ============================================
# CELL 4 — 1D-CNN Deep Learning Model
# ============================================

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import numpy as np

# CNN expects 3D input: (samples, timesteps, features)
# Our features = 5, so we reshape from (N,5) → (N,5,1)
X_train_cnn = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test_cnn  = X_test.reshape((X_test.shape[0],  X_test.shape[1],  1))

# CNN Model
model = Sequential([
    Conv1D(32, kernel_size=2, activation='relu', input_shape=(5,1)),
    Conv1D(64, kernel_size=2, activation='relu'),
    MaxPooling1D(pool_size=2),
    Flatten(),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(3, activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train
history = model.fit(
    X_train_cnn, y_train,
    epochs=10,
    batch_size=64,
    validation_split=0.2,
    verbose=1
)

# Predict
y_pred_cnn = np.argmax(model.predict(X_test_cnn), axis=1)

# Evaluation
print("\nCNN Test Accuracy:", accuracy_score(y_test, y_pred_cnn))
print("\nClassification Report:")
print(classification_report(y_test, y_pred_cnn))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred_cnn)
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Greens",
            xticklabels=["Benign","Scan","CnC"],
            yticklabels=["Benign","Scan","CnC"])
plt.title("CNN Confusion Matrix")
plt.xlabel("Predicted"); plt.ylabel("Actual")
plt.show()
```

```python
# ============================================
# CELL 5 — Three binary modules (XGBoost + small DNN)
# Modules: Benign-vs-others, Scan-vs-others, CnC-vs-others
# ============================================

# Requirements: X_train, X_test, y_train, y_test, scaler must exist from previous cells

# 1) Imports
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Input
from tensorflow.keras.callbacks import EarlyStopping

# Utility to plot confusion matrix
def plot_cm(cm, title):
    plt.figure(figsize=(4,3))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
    plt.title(title)
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

# Map multi-class labels to binary for a target class
def make_binary_labels(y, target_class):
    # returns array: 1 for target_class, 0 otherwise
    return np.where(y == target_class, 1, 0)

# Keep track of trained models and outputs
modules = ["Benign", "Scan", "CnC"]
target_map = {"Benign":0, "Scan":1, "CnC":2}

xgb_modules = {}
dnn_modules = {}
probs_train = {}  # store module probabilities on train (for combiner later if needed)
probs_test = {}   # store module probabilities on test

# Use a small DNN builder function
def build_small_dnn(input_shape):
    model = Sequential([
        Input(shape=(input_shape,)),
        Dense(64, activation='relu'),
        Dropout(0.2),
        Dense(32, activation='relu'),
        Dropout(0.1),
        Dense(1, activation='sigmoid')  # binary output
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Training loop for each module
for mod in modules:
    print(f"\n\n===== Module: {mod} (binary) =====")
    tclass = target_map[mod]

    # Prepare binary labels
    y_train_bin = make_binary_labels(y_train, tclass)
    y_test_bin  = make_binary_labels(y_test, tclass)

    # --- XGBoost (fast, strong baseline) ---
    print("\n-> Training XGBoost...")
    xgb = XGBClassifier(
        n_estimators=150,
        max_depth=5,
        learning_rate=0.1,
        use_label_encoder=False,
        eval_metric='logloss',
        random_state=42
    )
    xgb.fit(X_train, y_train_bin)
    y_pred_xgb = xgb.predict(X_test)
    y_prob_xgb = xgb.predict_proba(X_test)[:,1]

    print("XGBoost metrics:")
    print(" Accuracy:", accuracy_score(y_test_bin, y_pred_xgb))
    print(classification_report(y_test_bin, y_pred_xgb, digits=4))
    cm_x = confusion_matrix(y_test_bin, y_pred_xgb)
    plot_cm(cm_x, f"XGB CM - {mod}")

    # Save
    xgb_modules[mod] = xgb
    probs_test[f"xgb_{mod}"] = y_prob_xgb

    # --- Small DNN (Keras) ---
    print("\n-> Training small DNN...")
    dnn = build_small_dnn(X_train.shape[1])
    # Early stopping to avoid long runs
    es = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True, verbose=0)
    dnn.fit(X_train, y_train_bin, epochs=30, batch_size=128, validation_split=0.15, callbacks=[es], verbose=0)

    y_prob_dnn = dnn.predict(X_test).flatten()
    y_pred_dnn = (y_prob_dnn >= 0.5).astype(int)

    print("DNN metrics:")
    print(" Accuracy:", accuracy_score(y_test_bin, y_pred_dnn))
    print(classification_report(y_test_bin, y_pred_dnn, digits=4))
    cm_d = confusion_matrix(y_test_bin, y_pred_dnn)
    plot_cm(cm_d, f"DNN CM - {mod}")

    # Save
    dnn_modules[mod] = dnn
    probs_test[f"dnn_{mod}"] = y_prob_dnn

# Summarize stored probability keys
print("\nStored module probability keys (test):", list(probs_test.keys()))
```

```python
# ============================================
# CELL 6 — Adaptive Modular Neural Network (AMNN) combiner (hybrid stacking)
# Uses module probabilities from XGBoost + DNN modules (from CELL 5)
# ============================================

# Required objects from earlier cells:
# - xgb_modules (dict with XGB models for 'Benign','Scan','CnC')
# - dnn_modules (dict with DNN models for same)
# - X_train, X_test, y_train, y_test  (scaled features)
# - (If missing, rerun Cell 5 to recreate)

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1) Build module probability features for a dataset selector (train/val/test)
def build_module_probs(X, xgb_models, dnn_models):
    """
    X: (N, F) scaled numeric features
    returns: (N, 6) array of probabilities:
    [xgb_benign, xgb_scan, xgb_cnc, dnn_benign, dnn_scan, dnn_cnc]
    """
    probs = []
    # XGBoost module probs: predict_proba -> we used binary tasks, so idx 1 = prob of positive (target)
    p_xgb_b = xgb_models["Benign"].predict_proba(X)[:,1]
    p_xgb_s = xgb_models["Scan"].predict_proba(X)[:,1]
    p_xgb_c = xgb_models["CnC"].predict_proba(X)[:,1]
    # DNN models: they output a single probability per sample
    p_dnn_b = dnn_models["Benign"].predict(X).flatten()
    p_dnn_s = dnn_models["Scan"].predict(X).flatten()
    p_dnn_c = dnn_models["CnC"].predict(X).flatten()
    probs = np.vstack([p_xgb_b, p_xgb_s, p_xgb_c, p_dnn_b, p_dnn_s, p_dnn_c]).T
    return probs

# Build meta-features for TRAIN and TEST
meta_X_train_full = build_module_probs(X_train, xgb_modules, dnn_modules)
meta_X_test = build_module_probs(X_test, xgb_modules, dnn_modules)
meta_y_train_full = y_train  # same labels

print("Meta-features shapes:", meta_X_train_full.shape, meta_X_test.shape)

# 2) Split meta training into train_meta and val_meta to tune/log performance
from sklearn.model_selection import train_test_split
meta_X_train, meta_X_val, meta_y_train, meta_y_val = train_test_split(
    meta_X_train_full, meta_y_train_full, test_size=0.20, random_state=42, stratify=meta_y_train_full
)
print("Meta train/val shapes:", meta_X_train.shape, meta_X_val.shape)

# 3) Baseline 1: simple argmax of XGBoost module probs -> predicted class by argmax of [p_b, p_s, p_c]
def argmax_from_xgb(X_meta):
    p_xgb_only = X_meta[:,0:3]  # first 3 are xgb probs
    preds = np.argmax(p_xgb_only, axis=1)  # 0->Benign,1->Scan,2->CnC
    return preds

# Baseline 2: argmax of DNN module probs
def argmax_from_dnn(X_meta):
    p_dnn_only = X_meta[:,3:6]
    preds = np.argmax(p_dnn_only, axis=1)
    return preds

# Evaluate baselines on test set
pred_xgb_baseline = argmax_from_xgb(meta_X_test)
pred_dnn_baseline = argmax_from_dnn(meta_X_test)

print("\nBaseline XGB-modules performance on test:")
print("Accuracy:", accuracy_score(y_test, pred_xgb_baseline))
print(classification_report(y_test, pred_xgb_baseline, digits=4))

print("\nBaseline DNN-modules performance on test:")
print("Accuracy:", accuracy_score(y_test, pred_dnn_baseline))
print(classification_report(y_test, pred_dnn_baseline, digits=4))

# 4) Train meta-classifier (Logistic Regression multinomial)
meta_clf = LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=2000, random_state=42)
meta_clf.fit(meta_X_train, meta_y_train)

# Evaluate on validation set
val_pred = meta_clf.predict(meta_X_val)
print("\nMeta-classifier (val) Accuracy:", accuracy_score(meta_y_val, val_pred))
print(classification_report(meta_y_val, val_pred, digits=4))

# Evaluate final on test set
meta_test_pred = meta_clf.predict(meta_X_test)
print("\nAMNN (meta) on test Accuracy:", accuracy_score(y_test, meta_test_pred))
print(classification_report(y_test, meta_test_pred, digits=4))

# Confusion matrix for meta-combiner
cm_meta = confusion_matrix(y_test, meta_test_pred)
plt.figure(figsize=(6,4))
sns.heatmap(cm_meta, annot=True, fmt='d', cmap='Oranges',
            xticklabels=["Benign","Scan","CnC"], yticklabels=["Benign","Scan","CnC"])
plt.title("AMNN (meta) Confusion Matrix")
plt.xlabel("Predicted"); plt.ylabel("Actual")
plt.show()

# 5) Save meta model (optional) and show feature importance (coeffs)
import joblib
joblib.dump(meta_clf, "/content/amnn_meta_clf.joblib")
print("Saved meta-classifier to /content/amnn_meta_clf.joblib")

# Show weights learned (logistic regression coefficients)
coef = meta_clf.coef_  # shape (3,6)
print("\nMeta-classifier coefficients (rows = class 0/1/2, cols = xgb_b,xgb_s,xgb_c,dnn_b,dnn_s,dnn_c):")
print(coef)

# 6) Quick comparison table
from sklearn.metrics import f1_score, precision_score, recall_score
def summarize(name, y_true, y_pred):
    return {
        "name": name,
        "accuracy": accuracy_score(y_true, y_pred),
        "f1_macro": f1_score(y_true, y_pred, average='macro'),
        "precision_macro": precision_score(y_true, y_pred, average='macro'),
        "recall_macro": recall_score(y_true, y_pred, average='macro')
    }

summary = [
    summarize("XGB-module-argmax", y_test, pred_xgb_baseline),
    summarize("DNN-module-argmax", y_test, pred_dnn_baseline),
    summarize("AMNN-meta", y_test, meta_test_pred)
]

import pandas as pd
print("\nSummary comparison:")
pd.DataFrame(summary)
```

```python
# ============================================
# CELL 8 — Unified AMNN Inference Pipeline
# ============================================

import joblib
import numpy as np
import tensorflow as tf

# Paths to saved models
BASE = "/content/saved_models"

# Load scaler
scaler_loaded = joblib.load(f"{BASE}/scaler/standard_scaler.pkl")

# Load XGB modules
xgb_benign = joblib.load(f"{BASE}/xgb_modules/xgb_Benign.pkl")
xgb_scan   = joblib.load(f"{BASE}/xgb_modules/xgb_Scan.pkl")
xgb_cnc    = joblib.load(f"{BASE}/xgb_modules/xgb_CnC.pkl")

# Load DNN modules
dnn_benign = tf.keras.models.load_model(f"{BASE}/dnn_modules/dnn_Benign.h5")
dnn_scan   = tf.keras.models.load_model(f"{BASE}/dnn_modules/dnn_Scan.h5")
dnn_cnc    = tf.keras.models.load_model(f"{BASE}/dnn_modules/dnn_CnC.h5")

# Load meta-classifier
meta_clf_loaded = joblib.load(f"{BASE}/amnn/meta_classifier.pkl")

print("All models loaded successfully!")


# ----------------------------------------------------
# FINAL AMNN PREDICTION FUNCTION
# ----------------------------------------------------
def amnn_predict(flow_dict):
    """
    flow_dict must contain the 5 numerical features:
      duration, orig_bytes, resp_bytes, orig_pkts, resp_pkts
    Example:
      {"duration": 1.2, "orig_bytes": 50, "resp_bytes": 300,
       "orig_pkts": 10, "resp_pkts": 12}
    Returns:
      final_class (0=Benign, 1=Scan, 2=CnC)
    """

    # 1) Convert dict → numpy array
    features = np.array([
        flow_dict["duration"],
        flow_dict["orig_bytes"],
        flow_dict["resp_bytes"],
        flow_dict["orig_pkts"],
        flow_dict["resp_pkts"]
    ]).reshape(1, -1)

    # 2) Scale
    X_scaled = scaler_loaded.transform(features)

    # 3) Get probabilities from XGB modules
    p_xgb_b = xgb_benign.predict_proba(X_scaled)[:,1]
    p_xgb_s = xgb_scan.predict_proba(X_scaled)[:,1]
    p_xgb_c = xgb_cnc.predict_proba(X_scaled)[:,1]

    # 4) Get probabilities from DNN modules
    p_dnn_b = dnn_benign.predict(X_scaled).flatten()
    p_dnn_s = dnn_scan.predict(X_scaled).flatten()
    p_dnn_c = dnn_cnc.predict(X_scaled).flatten()

    # 5) Build meta-feature vector
    meta_vec = np.array([
        p_xgb_b[0], p_xgb_s[0], p_xgb_c[0],
        p_dnn_b[0], p_dnn_s[0], p_dnn_c[0]
    ]).reshape(1, -1)

    # 6) Meta-classifier final prediction
    final_class = meta_clf_loaded.predict(meta_vec)[0]

    return final_class


# ----------------------------------------------------
# TEST THE FINAL PIPELINE WITH A SAMPLE INPUT
# ----------------------------------------------------
sample_flow = {
    "duration": 1.5,
    "orig_bytes": 100,
    "resp_bytes": 200,
    "orig_pkts": 5,
    "resp_pkts": 7
}

print("Sample Prediction:", amnn_predict(sample_flow))
```
